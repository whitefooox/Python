```python
fig, ax = plt.subplots()

ax.set_title("Closing stock value of Alphabet Inc.")
ax.set_xlabel("Date")
ax.set_ylabel("Closing Value")

values = [772.5, 776.4, 776.5, 776.8, 775.1]
dates = [
    datetime(2016,10,3),
    datetime(2016,10,4),
    datetime(2016,10,5),
    datetime(2016,10,6),
    datetime(2016,10,7),
]

ax.set_xlim(datetime(2016,10,3), datetime(2016,10,7))
ax.set_ylim(772.5, 777.0)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
ax.xaxis.set_major_formatter(DateFormatter("%Y-%m-%d"))
ax.xaxis.set_major_locator(mdates.DayLocator())
ax.tick_params(which='major', labelsize=10)
ax.plot(dates, values, c='r', linewidth=0.8)
ax.scatter(dates, values, edgecolor='black', color='r')
plt.show()
```

```python
X = np.linspace(0.5, 3.5, 100)
Y1 = 3 + np.cos(X)
Y2 = 1 + np.cos(1 + X / 0.75) / 2
Y3 = np.random.uniform(Y1, Y2, len(X))

fig = plt.figure(figsize=(7.5, 7.5))
ax = fig.add_axes([0.2, 0.17, 0.68, 0.7], aspect=1)

ax.xaxis.set_major_locator(MultipleLocator(1.000))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_major_locator(MultipleLocator(1.000))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
ax.xaxis.set_minor_formatter("{x:.2f}")
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

ax.tick_params(which='major', width=1.0, length=10, labelsize=14)
ax.tick_params(which='minor', width=1.0, length=5, labelsize=10,
               labelcolor='0.25')

ax.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10)

ax.plot(X, Y1, c='b', lw=2.5, label=" ", zorder=10)
ax.plot(X, Y2, c='r', lw=2.5, label=" ")
ax.plot(X[::3], Y3[::3], linewidth=0, markersize=7,
        marker='o', markerfacecolor='none', markeredgecolor='black',
        markeredgewidth=2.5)

ax.set_title(" ", fontsize=20, verticalalignment='bottom')
ax.set_xlabel("  OX", fontsize=14)
ax.set_ylabel("  OY", fontsize=14)
ax.legend(loc="upper right", fontsize=14)

plt.show()
```

```python
fig, ax = plt.subplots()

X = np.arange(-10, 10, 0.01)
Y = lambda X: np.log((X * X + 1) * np.exp(-np.absolute(X) / 10)) / np.log(1 + np.tan(1 / (1 + np.sin(X) * np.sin(X))))

ax.plot(X, Y(X))
plt.show()
```

```python
n_groups = 5
men_means = (22, 30, 33, 30, 26)
women_means = (25, 32, 30, 35, 29)

fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, men_means, bar_width,
alpha=opacity,
color='g',
label='Men')

rects2 = plt.bar(index + bar_width, women_means, bar_width,
alpha=opacity,
color='r',
label='Women')

plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + bar_width, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend()

plt.tight_layout()
plt.show()
```
